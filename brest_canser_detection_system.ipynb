```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('datasets/breast-cancer.csv')
```

```python
df.head()
```

```python
df.drop(columns=['id'],inplace= True)
```

```python
x=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
col=x.columns.tolist()
```

```python
processing=ColumnTransformer(
    transformers = [
        ('standardize',StandardScaler(),col)
    ]
)
```

```python
xtrains=processing.fit_transform(x_train)
xtests=processing.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier,VotingClassifier,BaggingClassifier,StackingClassifier
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
decision_tree =DecisionTreeClassifier()
random_forest = RandomForestClassifier()
adaboost = AdaBoostClassifier()
knn = KNeighborsClassifier()
gradientboosting = GradientBoostingClassifier()
voting = VotingClassifier(estimators=[('lr', LogisticRegression()), ('dt', DecisionTreeClassifier()), ('km', knn)])
bagging = BaggingClassifier()
stacking = StackingClassifier(estimators=[('lr', LogisticRegression()), ('dt', DecisionTreeClassifier()), ('km', knn)])
```

```python
from sklearn.metrics import accuracy_score
```

```python
def train_test(name,model,x_train,x_test,y_train,y_test):
    print(f"{name} is training...\n")
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    print(f"accuracy_score of {name} is : {accuracy_score(y_test,pred)} \n\n")
```

```python
models=[decision_tree,random_forest,adaboost,knn,gradientboosting,voting,bagging,stacking]
names=['decision_tree','random_forest','adaboost','knn','gradientboosting','voting','bagging','stacking']
```

```python
for i in range(len(models)):
    train_test(names[i],models[i],xtrains,xtests,y_train,y_test)
```

```python
model = Pipeline(steps =[
    ('processing',processing),
    ('stacking',stacking)
])
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
accuracy_score(y_test,pred)
```

```python
import joblib
```

```python
joblib.dump(model,'breast_cancer_predicter.joblib')
```
